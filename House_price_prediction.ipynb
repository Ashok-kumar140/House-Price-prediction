```python
# Implementing the Multi linear model for predicting profit.
```

```python
x = new_df.iloc[:,0:4].values
```

```python
y = new_df.iloc[:,-1].values
y
```

```python
# Spliting our data into train and test data sets.
```

```python
from sklearn.model_selection import train_test_split   
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1)             
```

```python
#By default it split our data into 0.75 and 0.25 ratio.
```

```python
# importing LinearRegression model.

from sklearn.linear_model import LinearRegression 

#Model calling
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# Predicted data by our model
y_pred = model.predict(x_test)
y_pred   
```

```python
# Actual data
y_test
```

```python
# Here we can visualize our predicted and actual data side by side.
dataframe = {'Predicted': y_pred, "Actual": y_test}
data = pd.DataFrame(dataframe)
data
```

```python
# Here i am calculating accuracy and mean square error and root mean square error of our model.
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# r2_score
score=r2_score(y_test,y_pred)
print("Accuracy score : ", score)
```

```python
# Mean square error
mean_sqrd_error = mean_squared_error(y_test,y_pred)
print("Mean square error is : ",mean_sqrd_error)
```

```python
# Root mean square error
import numpy as np
root_mean_sqrd_error = np.sqrt(mean_sqrd_error)
print("Root mean square error is : ", root_mean_sqrd_error)
```
